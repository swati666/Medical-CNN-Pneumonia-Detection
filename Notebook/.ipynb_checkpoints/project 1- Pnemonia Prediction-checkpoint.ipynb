```python
import tensorflow as tf
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_gen = ImageDataGenerator(
    rescale= 1./255,
    rotation_range = 15,
    zoom_range = 0.1,
    horizontal_flip = True
)
```

```python
val_gen = ImageDataGenerator(rescale=1./255)
```

```python
train_data = train_gen.flow_from_directory(
    r"D:\ML Projects\chest_xray\train",
    target_size = (224,224),
    #color_mode = "grayscale",
    color_mode = 'rgb',
    batch_size = 32,
    class_mode = "binary"    
)

val_data = val_gen.flow_from_directory(
    r"D:\ML Projects\chest_xray\val",
    target_size = (224,224),
    #color_mode = "grayscale",
    color_mode = 'rgb',
    batch_size = 32,
    class_mode = "binary"
)

test_data = val_gen.flow_from_directory(
    r"D:\ML Projects\chest_xray\test",
    target_size = (224,224),
    #color_mode = "grayscale", # 
    color_mode = 'rgb',
    batch_size = 32,
    class_mode = "binary",
    shuffle = False
    
)
```

```python
print(train_data.class_indices)
```

```python
from tensorflow.keras import layers,models
```

```python
model = models.Sequential([
    layers.Conv2D(
        filters = 32,
        kernel_size = (3,3),
        activation = 'relu',
        padding = 'same',
        input_shape = (224,224,1)
    ),
    layers.MaxPooling2D(2,2),

    #Block 2
    layers.Conv2D(
        filters = 64,
        kernel_size = (3,3),
        activation = 'relu',
        padding = 'same'
    ),
    layers.MaxPooling2D(2,2),

    #Block 3 
    layers.Conv2D(
        filters = 128,
        kernel_size = (3,3),
        activation = 'relu',
        padding = 'same'
    ),
    layers.MaxPooling2D(2,2),

    #Classifier 
    layers.Flatten(),
    layers.Dense(128,activation = 'relu'),
    layers.Dropout(0.5),
    layers.Dense(1,activation='sigmoid')

])
                  
                  
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy', tf.keras.metrics.Recall()]
)
```

```python
history = model.fit(
    train_data,
    epochs = 15,
    validation_data = val_data
)
```

```python
plt.figure(figsize = (12,4))

plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label="Acuuracy")
plt.plot(history.history['val_accuracy'], label = "Validation Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Echos")
plt.legend()
plt.title("Accuracy Plot")

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label = "Loss")
plt.plot(history.history['val_loss'], label = "Validation Loss")
plt.ylabel("Loss")
plt.xlabel("Echos")
plt.legend()
plt.title("Loss Plot")

plt.show()
```

```python
test_loss, test_accuracy, test_recall = model.evaluate(test_data)

print("Test Accuracy", test_accuracy)
print("Test Recall", test_recall)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
```

```python
y_pred = model.predict(test_data)
y_pred = (y_pred >0.5).astype(int)

cm = confusion_matrix(test_data.classes, y_pred)

print(cm)
print(classification_report(test_data.classes, y_pred))
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
```

```python
base_model = MobileNetV2(
    weights="imagenet",
    include_top=False,
    input_shape=(224,224,3)
)

base_model.trainable = False
```

```python
transfer_model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(128,activation = 'relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation = 'sigmoid')
])
```

```python
transfer_model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy', tf.keras.metrics.Recall()]
)
```

```python
tranfer_history = transfer_model.fit(
    train_data,
    validation_data = val_data,
    epochs = 10   
)
```

```python
plt.figure(figsize = (12,4))

plt.subplot(1,2,1)
plt.plot(tranfer_history.history['accuracy'], label = "Accuracy")
plt.plot(tranfer_history.history['val_accuracy'], label = "Val_Accuracy")
plt.title("Accuracy Plot")
plt.xlabel("Echos")
plt.ylabel("Accuracy")
plt.legend()


plt.subplot(1,2,2)
plt.plot(tranfer_history.history['loss'], label = 'loss')
plt.plot(tranfer_history.history['val_loss'], label = 'Val Loss')
plt.title("Loss Plot")
plt.xlabel("Echos")
plt.ylabel("Loss")
plt.legend()

plt.show()
```

```python
transfer_model.evaluate(test_data)
```
